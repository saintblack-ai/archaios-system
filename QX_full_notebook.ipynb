## Detailed Chapter Synopses

### Quantum Computation and Quantum Information (Nielsen & Chuang, 2010)
1. **Introduction**: Motivation, classical vs. quantum paradigms.
2. **Quantum Mechanics and Notation**: Postulates; Dirac notation; tensor products.
3. **Quantum Computation: Basic Concepts**: Qubits; Bloch sphere; gates; circuits.
4. **Quantum Algorithms**: Deutsch–Jozsa; Simon’s; Shor’s algorithm.
5. **Quantum Search and Amplitude Amplification**: Grover’s algorithm.
6. **Quantum Information Theory**: Entropy; channel capacities.
7. **Quantum Error Correction**: Shor code; CSS codes.
8. **Fault-Tolerant Quantum Computation**: Threshold theorems.
9. **Physical Implementations**: Ion traps; superconducting circuits.
10. **Outlook and Advanced Topics**: Topological computing; cluster states.

### Probabilistic Robotics (Thrun, Burgard & Fox, 2005)
1. **Introduction to Uncertainty**: Sources; probabilistic models.
2. **Kalman Filters**: Linear Gaussian estimation.
3. **Extended/Unscented Kalman Filters**: Nonlinear models.
4. **Particle Filters**: Global localization.
5. **SLAM**: FastSLAM; GraphSLAM.
6. **Control**: MDPs; POMDPs.
7. **Planning under Uncertainty**: PRMs; belief-space planning.
8. **Applications**: Mobile robots; autonomous vehicles.

### The Physics of Nuclear Reactors (Marguet, 2017)
1. **Neutron Transport Theory**: Boltzmann equation.
2. **Multigroup Diffusion**: Energy discretization.
3. **Reactor Kinetics and Dynamics**: Point kinetics.
4. **Thermal–Hydraulics Coupling**: Coolant flow.
5. **Monte Carlo Methods**: Variance reduction.
6. **Core Design and Optimization**: Fuel assemblies.
7. **Safety Analysis and PRA**: Risk assessment.
8. **Advanced Reactor Concepts**: SMRs; molten salt reactors.

## Code Samples

### Quantum Computing

```
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0,1], [0,1])

backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1024)
counts = job.result().get_counts()
print("Measurement counts:", counts)
plot_histogram(counts)
```

### Robotics

```
import rclpy
from rclpy.node import Node
from std_msgs.msg import String

class HelloPublisher(Node):
    def __init__(self):
        super().__init__('hello_publisher')
        self.pub = self.create_publisher(String, 'greetings', 10)
        self.timer = self.create_timer(1.0, self.timer_callback)
        self.count = 0

    def timer_callback(self):
        msg = String()
        msg.data = f'Hello, world! {self.count}'
        self.pub.publish(msg)
        self.get_logger().info(f'Published: "{msg.data}"')
        self.count += 1

def main(args=None):
    rclpy.init(args=args)
    node = HelloPublisher()
    rclpy.spin(node)
    node.destroy_node()
    rclpy.shutdown()

if __name__ == '__main__':
    main()
```

### Nuclear Energy

```
import numpy as np
import matplotlib.pyplot as plt

L, N = 100.0, 101
dx = L/(N-1)
D, Sigma_a, nuSigma_f = 1.2, 0.01, 0.02

A = np.zeros((N, N))
S = np.zeros(N)
for i in range(1, N-1):
    A[i,i-1] = D/dx**2
    A[i,i]   = -2*D/dx**2 - Sigma_a
    A[i,i+1] = D/dx**2
    S[i]    = -nuSigma_f

A[0,0] = A[-1,-1] = 1.0
phi = np.linalg.solve(A, -S)

plt.plot(np.linspace(0, L, N), phi)
plt.xlabel('Position (cm)')
plt.ylabel('Neutron Flux')
plt.title('1D Neutron Diffusion')
plt.show()
```
